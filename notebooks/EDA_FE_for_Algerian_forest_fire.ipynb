##Algerian forest fires dataset

Information:

The dataset includes 244 instances that regroup a data of two regions of algeria namely the bejaia region located in the northeast of algeria and the sidi bel-abbes region located in the north west  of algeria.

122 instances for each region
The period from June 2012 to September 2012 the dataset includes 11 attributes and 1 output attribute(class) the 244 instances have been classified into fire(138 classes) and not fire(106 classes) classes .


#Attribute information:

1. date-> (DD/MM/YYYY) dat , month(‘june ‘ to ‘september’), year(2012) weather data observations
2. Temp : temperature noon(temperature max) in celsius degrees 22 to 42
3. RH : relative humidity in % 21 to 90
4. ws : wind speed in km./h 6to 29
5. RAIN : total day in mm 0 to 16.8 FWI components.
6. fine fuel moisture code (FFMC) indes from the fwi system 28.6 to 92.
7. duff moisture code(dmc) index from fwi system 1.1 to 65.9
8. drought code(DC) index from fwi system 7 to 220.4
9. initial spread index (ISI) index from the fwi system 0 to 18.5
10. buildup index (BUI) index from the fwi system  1.1 to 68
11. fire weather index(fwi) index 0 to 31.1
12. classes two classes: namely fire and not fire.


The dataset is converted into two sets based on region from 122th  index , we make a new column based on the region,

1: “Bejaia Region Dataset”

2: “Sidi-Bel Abbes Region Dataset”

Add new column with region


```python
df.loc[0:122,"Region"] = 0
df.loc[122:,"Region"] = 1
```

```python
df.info()
```

```python
# the newley created column is of float type but i need that data int integer format
df['Region'] = df['Region'].astype(int)
```

```python
#just dropping the null values
df = df.dropna().reset_index(drop=True)
```

```python
#fixing the column name
df.columns = df.columns.str.strip()
```

```python
df.columns
```

```python
#convert this columns from object int data type
convert_column = ['day', 'month', 'year', 'Temperature', 'RH', 'Ws']

for column in convert_column:
  df[column] = df[column].astype(int)
```

```python
# #converting the remaining columns which as object is converted into float

objects = [features for features in df.columns if df[features].dtypes == 'O']

```

```python
objects
```

```python
for i in objects:
  if i!='Classes':
     df[i] =df[i].astype(float)
```
